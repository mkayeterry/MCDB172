```python
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import numpy as np

# Get the RGB values of the color
rgb = mcolors.to_rgb('#3b528b')

# Adjust the blue value to make it slightly more blue
new_rgb = (rgb[0], rgb[1], 0.6)

# Convert back to hex code
new_color = mcolors.to_hex(new_rgb)

x = np.array([1, 10])
y = np.array([10, 1])
plt.plot(x, y, linewidth=10, color = new_color)
```
